```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
```

## 4 PARAMETERS

```python
files = glob.glob("/Users/ccristiano/Documents/Codigos/QML/MNIST/qutrit/data/TestConvergence/train_progress_NUMFEATURES_6_NUM_PARAMS_4_*.csv")
files_sorted = sorted(files, key=lambda x: int(os.path.splitext(os.path.basename(x))[0].split("_")[-1]))

dfs = []
for file in files_sorted:
    df = pd.read_csv(file)
    # opcional: guardar nmero de trial segn el nombre
    trial_num = int(os.path.splitext(os.path.basename(file))[0].split("_")[-1])
    df["FileTrial"] = trial_num
    dfs.append(df)
all_data = pd.concat(dfs, ignore_index=True)

# plt.figure(figsize=(8, 4))
# plt.plot(all_data["Trial"], all_data["BestTrainLoss"], marker='o', label="Loss")
# plt.xlabel("Trial")
# plt.ylabel("Best Train Loss so far")
# plt.grid(True)
# plt.tight_layout()
# plt.show()

# Average every 200 trials
all_data["trial_block"] = all_data["Trial"] // 200
loss_means = all_data.groupby("trial_block")["BestTrainLoss"].mean().reset_index()
plt.figure(figsize=(8,4))
plt.plot(loss_means["trial_block"], loss_means["BestTrainLoss"], marker='o', color='indianred')
plt.xlabel("Trial Block (each = 200 trials)", fontsize=17)
plt.ylabel("Average Best Train Loss", fontsize=17)
plt.title("Features = 5 | Params = 4", fontsize=17)
plt.grid(linestyle='dotted', color='lightgrey')
plt.show()

BestLoss_4params = min(all_data["BestTrainLoss"])
BestTrial_4params = all_data["Trial"][all_data["BestTrainLoss"] == BestLoss_4params].iloc[0]
print("Best Loss = ", str(BestLoss_4params))
print("Best Trial = ", str(BestTrial_4params))
```
