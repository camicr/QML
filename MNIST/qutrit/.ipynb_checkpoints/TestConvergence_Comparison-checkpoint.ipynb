```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib as mpl

# Configuración general para gráficos estilo publicación
mpl.rcParams.update({
    # Fuente estilo LaTeX
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern"],

    # Tamaño de fuente general
    "font.size": 12,
    
    # Tamaños para títulos, etiquetas y ticks
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,

    # Grosor de líneas y ejes
    "lines.linewidth": 1.5,
    "axes.linewidth": 1.0,
    
    # Tamaño de marcadores
    "lines.markersize": 6,

    # Tamaño de figuras
    "figure.figsize": (6, 4),  # En pulgadas

    # Espaciado entre etiquetas y ticks
    "xtick.major.size": 4,
    "ytick.major.size": 4,

    # Estilo de leyenda
    "legend.fontsize": 10,
    "legend.frameon": False,

    # Fondo blanco
    "figure.facecolor": "white",
    "axes.facecolor": "white"
})
```

```python
files = ["/Users/ccristiano/Documents/Codigos/QML/MNIST/qutrit/data/TestConvergence/data_NUM_FEATURES_4.csv",
         "/Users/ccristiano/Documents/Codigos/QML/MNIST/qutrit/data/TestConvergence/data_NUM_FEATURES_5.csv",
         "/Users/ccristiano/Documents/Codigos/QML/MNIST/qutrit/data/TestConvergence/data_NUM_FEATURES_6.csv",
         "/Users/ccristiano/Documents/Codigos/QML/MNIST/qutrit/data/TestConvergence/data_NUM_FEATURES_7.csv",
         "/Users/ccristiano/Documents/Codigos/QML/MNIST/qutrit/data/TestConvergence/data_NUM_FEATURES_8.csv"]

Best_Loss = []
n_params  = []
for fi in files:
    dfi = pd.read_csv(fi)
    Best_Loss.append(dfi["Best_Loss"])
    n_params.append(dfi["n_params"])
```

```python
n_features = [4, 5, 6, 7, 8]
# palette = sns.color_palette("hls", len(n_features))
palette = sns.color_palette("rocket", len(n_features))

plt.figure(figsize=(8,4))
for j in range(len(Best_Loss)):
    plt.plot(np.array(n_params[j]), np.array(Best_Loss[j]), color=palette[len(palette)-1-j], marker='o', label=str(n_features[j]))
plt.xlabel("# of params", fontsize=17)
plt.ylabel("Best Loss", fontsize=17)
plt.xticks(n_params[-1])
plt.grid(linestyle='dotted', color='lightgrey')
plt.legend(title='# of features', fontsize=12, title_fontsize=12)
plt.title("4000 initial points", fontsize=17)
plt.tight_layout()
# plt.savefig("/Users/ccristiano/Downloads/plot.png")
plt.show()
```
