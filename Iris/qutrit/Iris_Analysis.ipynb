## 3 features, 3 parameters (C30) (4000 puntos iniciales ✓)

$$U(\vec{x}) = \exp(ix_1\lambda_1 + ix_1\lambda_2 + ix_1\lambda_3)$$ 
$$U(\vec{\theta}) = \exp(i\theta_1\lambda_1 + i\theta_1\lambda_2 + i\theta_1\lambda_3)$$

(and then, sequentially add parameters)

```python
import glob, pandas as pd
import matplotlib.pyplot as plt

BASE_SAVE_DIR = "/Users/ccristiano/Documents/Codigos/QML/Iris/qutrit/data/TestConvergence_SequentialTraining"
NUM_FEATURES = 3
NUM_PARAMS   = 3
LAYERS       = 6
CONFIG       = 30

files = sorted(glob.glob(f"{BASE_SAVE_DIR}/train_progress_NUMFEATURES_{NUM_FEATURES}_NUM_PARAMS_{NUM_PARAMS}_LAYERS_{LAYERS}_{CONFIG}*.csv"))
df = pd.concat((pd.read_csv(f) for f in files), ignore_index=True)
df = df.sort_values("Trial").reset_index(drop=True)

BestLoss = df["BestTrainLoss"].iloc[-1]
print(f"Best train loss: {BestLoss}")

BestTrainAcc = df["BestTrainAcc"].iloc[-1]
print(f"Best train loss: {BestTrainAcc}")

# Plot
plt.figure(figsize=(10,4))
plt.plot(df["Trial"], df["BestTrainLoss"].cummin(), drawstyle="steps-post", linewidth=2)
plt.xlabel("Trial")
plt.ylabel("Best Train Loss (so far)")
plt.title(f"Convergencia por trials (NUM_FEATURES={NUM_FEATURES}, NUM_PARAMS={NUM_PARAMS})")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(df["Trial"], df["BestTrainAcc"], drawstyle="steps-post", linewidth=2)
plt.xlabel("Trial")
plt.ylabel("Train Accuracy")
plt.title(f"Convergencia por trials (NUM_FEATURES={NUM_FEATURES}, NUM_PARAMS={NUM_PARAMS})")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

#Save data
# filename = "Comparacion.csv"
# row = {
#     "Config": CONFIG,
#     "BestLoss": df["BestTrainLoss"].iloc[-1],
#     "TrainAcc": df["BestTrainAcc"].iloc[-1]
# }
# file_exists = os.path.isfile(filename)

# with open(filename, mode="a", newline="") as f:
#     writer = csv.DictWriter(f, fieldnames=["Config", "BestLoss", "TrainAcc"])
#     if not file_exists:
#         writer.writeheader()  
#     writer.writerow(row)
```

# Comparación

```python
import pandas as pd

filename = "Comparacion.csv"

df = pd.read_csv(filename)

labels   = df["Config"].astype(str).iloc[1:]
BestLoss = df["BestLoss"].iloc[1:]
TrainAcc = df["TrainAcc"].iloc[1:]

# BestLoss
fig, ax = plt.subplots()
ax.plot(labels, BestLoss, marker="o", linestyle="-")
for x, y in zip(range(len(labels)), BestLoss):
    ax.text(x, y, labels.iloc[x], ha="center", va="bottom")  
ax.set_xlabel("CONFIG")
ax.set_ylabel("BestLoss")
plt.show()

# TrainAcc
fig, ax = plt.subplots()
ax.plot(labels, TrainAcc, marker="o", linestyle="-", color="green")
for x, y in zip(range(len(labels)), TrainAcc):
    ax.text(x, y, labels.iloc[x], ha="center", va="bottom")
ax.set_xlabel("CONFIG")
ax.set_ylabel("TrainAcc")
plt.show()
```
